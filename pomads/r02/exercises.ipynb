```python
from scipy.stats import binom

n = 137
p = 0.4

min50 = 0

for k in range(n + 1):
    prawdopodobienstwo = binom.pmf(k, n, p)
    min50 += prawdopodobienstwo if k >= 50 else 0
    # min50 += prawdopodobienstwo
    print("{0} - {1}".format(k, prawdopodobienstwo))

print(min50)
```

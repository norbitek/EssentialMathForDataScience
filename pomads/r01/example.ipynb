```python
def pochodna_x(f, x, rozmiar_kroku):
    m = (f(x + rozmiar_kroku) - f(x)) / ((x + rozmiar_kroku) - x)
    return m

def moja_funkcja(x):
    return x**2

nachylenie_w_punkcie_2 = pochodna_x(moja_funkcja, 2, .00001)
print(nachylenie_w_punkcie_2) # wypisuje 4.000010000000827
```

```python
from sympy import *
# Deklarujemy 'x' w SymPy
x = symbols('x')
# Teraz uywamy zwykej skadni Pythona, aby zadeklarowa funkcj
f = x**2
# Obliczamy pochodn funkcji
dx_f = diff(f)
print(dx_f) # wypisuje 2*x

dx_f.subs(x, 2)
```

```python
from sympy import *
from sympy.plotting import plot3d
# Deklarujemy x i y w SymPy
x,y = symbols('x y')
# Teraz uywamy zwykej skadni Pythona do zadeklarowania funkcji
f = 2*x**3 + 3*y**3
# Obliczamy pochodne czstkowe wzgldem x i y
dx_f = diff(f, x)
dy_f = diff(f, y)
print(dx_f) # wypisuje 6*x**2
print(dy_f) # wypisuje 9*y**2
# Tworzymy wykres funkcji
plot3d(f)
```

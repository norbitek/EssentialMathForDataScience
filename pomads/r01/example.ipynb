```python
def pochodna_x(f, x, rozmiar_kroku):
    m = (f(x + rozmiar_kroku) - f(x)) / ((x + rozmiar_kroku) - x)
    return m

def moja_funkcja(x):
    return x**2

nachylenie_w_punkcie_2 = pochodna_x(moja_funkcja, 2, .00001)
print(nachylenie_w_punkcie_2) # wypisuje 4.000010000000827
```

```python
from sympy import *

# "x" i wielko przyrostu "s"
x, s = symbols('x s')

# deklarujemy funkcj
f = x**2

# nachylenie linii biegncej midzy dwoma punktami oddalonymi o "s"
# uywamy wzoru "rnica wysokoci w pionie przez rnic odlegoci w poziomie"
nachylenie_f = (f.subs(x, x + s) - f) / ((x+s) - x)

# podstawiamy 2 za x
nachylenie_2 = nachylenie_f.subs(x, 2)

# obliczamy nachylenie w punkcie x = 2
# w nieskoczono zmniejszamy przyrost _s_ tak, aby zblia si do 0
wynik = limit(nachylenie_2, s, 0)

print(wynik) # 4
```

```python
#regua acuchowa

from sympy import *
x, y = symbols('x y')
# pochodna pierwszej funkcji
# poprzedzamy y znakiem podkrelenia, aby zapobiec konfliktowi zmiennych
_y = x**2 + 1
dy_dx = diff(_y)
# pochodna drugiej funkcji
z = y**3 - 2
dz_dy = diff(z)
# obliczamy pochodn przy uyciu i bez uycia
# reguy acuchowej, podstawiajc funkcj y
dz_dx_z_regula = (dy_dx * dz_dy).subs(y, _y)
dz_dx_bez_reguly = diff(z.subs(y, _y))
# dowodzimy reguy acuchowej, pokazujc, e obie pochodne s rwne
print(dz_dx_z_regula) # 6*x*(x**2 + 1)**2
print(dz_dx_bez_reguly) # 6*x*(x**2 + 1)**2
```

```python
def przybliz_calke(a, b, n, f):
    delta_x = (b - a) / n
    suma = 0

    for i in range(1, n + 1):
        pkt_srodkowy = 0.5 * (2 * a + delta_x * (2 * i - 1))
        suma += f(pkt_srodkowy)

    return suma * delta_x

def moja_funkcja(x):
    return x**2 + 1

pole = przybliz_calke(a=0, b=1, n=5, f=moja_funkcja)
print(pole) # wypisuje 1.33
```

```python
from sympy import *

# Deklarujemy 'x' w SymPy
x = symbols('x')

# Teraz uywamy zwykej skadni Pythona do zadeklarowania funkcji
f = x**2 + 1

# Obliczamy cak funkcji wzgldem x
# w obszarze midzy x = 0 a 1
pole = integrate(f, (x, 0, 1))

print(pole) # wypisuje 4/3
```

```python
from sympy import *

# Deklarujemy zmienne w SymPy
x, i, n = symbols('x i n')

# Deklarujemy funkcj i zakres
f = x**2 + 1
dolny, gorny = 0, 1

# Obliczamy szeroko i wysoko kadego prostokta o indeksie i
delta_x = ((gorny - dolny) / n)
x_i = (dolny + delta_x * i)
print(x_i)
fx_i = f.subs(x, x_i)
print(fx_i)

# Iterujemy po wszystkich n prostoktach i sumujemy ich pola
liczba_prostokatow = Sum(delta_x * fx_i, (i, 1, n)).doit()
print(liczba_prostokatow)

# Obliczamy pole, zwikszajc liczb
# prostoktw n do nieskoczonoci
pole = limit(liczba_prostokatow, n, oo)

print(pole) # wypisuje 4/3
```
